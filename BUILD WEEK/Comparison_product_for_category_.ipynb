```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Conto il numero di prodotti per Store e Categoria
conteggio_categorie = df_comparison.groupby(["Store", "category"]).size().reset_index(name="Quantità")

# Trovo le categorie comuni tra i due store
categorie_comuni = conteggio_categorie.groupby("category")["Store"].nunique()
categorie_comuni = categorie_comuni[categorie_comuni > 1].index  # Seleziona solo quelle presenti in entrambi i negozi

# Filtro il DataFrame mantenendo solo le categorie comuni
df_filtrato = conteggio_categorie[conteggio_categorie["category"].isin(categorie_comuni)]

# Creo una lista delle categorie comuni e dei negozi unici
categorie = df_filtrato["category"].unique()
stores = df_filtrato["Store"].unique()

# Posizionamento barre
x = np.arange(len(categorie))  # Posizioni sull'asse X
larghezza_barra = 0.35  # Larghezza delle barre

#Inserire uno style, se si vuole, visualizzati nella cella precedente. esempio: plt.style.use('ggplot')

plt.figure(figsize=(10, 6))

# Creo barre affiancate per ogni store
for i, store in enumerate(stores):
    valori = []
    
    # Estraggo le quantità per store e categoria
    for categoria in categorie:
        valore = df_filtrato[
            (df_filtrato["Store"] == store) &
            (df_filtrato["category"] == categoria)
        ]["Quantità"].sum()  # Sommo se ci sono più valori
        valori.append(valore)
    
    # Aggiungo delle barre al grafico
    plt.bar(x + i * larghezza_barra, valori, width=larghezza_barra, label=store)  


# etichette
plt.xticks(x + larghezza_barra / 2, categorie, rotation=45, ha="right")


plt.xlabel("Categoria di Prodotto")
plt.ylabel("Quantità di Prodotti")
plt.title("Quantità di prodotti per categoria nei due store (Solo categorie comuni)")
plt.legend(title="Store")
plt.grid(axis="y", linestyle="--", alpha=0.7) #alpha: opacità colore #linestyle: tratteggiate per una visualizzazione non pesante

# nota che il paramentro: linestyle, per alcuni stile, non è possibile settarlo,
# anzi in alcuni casi, elimina del tutto le righe del parametro dal grafico.

#show: per avere un output più pulito, senza avere ciò che è contentuto negli assi prima del grafico
plt.show()  
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Conto il numero di prodotti per Store e Categoria
conteggio_categorie = df_comparison[df_comparison["sex"] == "Donna"].groupby(["Store", "category"]).size().reset_index(name="Quantità")

# Trovo le categorie comuni tra i due store
categorie_comuni = conteggio_categorie.groupby("category")["Store"].nunique()
categorie_comuni = categorie_comuni[categorie_comuni > 1].index  # Seleziona solo quelle presenti in entrambi i negozi

# Filtro solo i prodotti da donna e con categorie comuni
df_donna = df_comparison[df_comparison["category"].isin(categorie_comuni)]

# Conto le quantità di prodotti per categoria e negozio
conteggio = df_donna.groupby(["category", "Store"]).size().unstack(fill_value=0)

# Creo grafico a barre affiancate
plt.figure(figsize=(10, 6))
larghezza_barra = 0.4  # Larghezza delle barre
x = np.arange(len(conteggio))  # Posizioni sull'asse X

# Barre per ogni store
plt.bar(x + larghezza_barra/2, conteggio["CafèNoir"], width=larghezza_barra, label="CafèNoir")#, color="green")
plt.bar(x - larghezza_barra/2, conteggio["NeroGiardini"], width=larghezza_barra, label="NeroGiardini")#, color="blu")
# segno + e - dopo la x danno precedenza su quale barra visualizzare prima da sinistra a destra dello schermo 

plt.xticks(x, conteggio.index, rotation=45)
plt.xlabel("Categorie")
plt.ylabel("Quantità di Prodotti")
plt.title("Quantità di prodotti per categoria (Donna)")
plt.legend(title="Store")
plt.grid(axis="y", linestyle="dotted", alpha=0.7) #linestyle: dotted (puntini)

plt.show()
```
